```python
import pandas as pd

import os
```

importing required libraries

```python
df = pd.read_csv('nifty_it_index.csv')
```

loaded dataset


```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

# Line Graph

```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='Close Price', color='blue')
plt.title('Nifty IT Index - Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Open'], label='Open Price', color='green')
plt.title('Nifty IT Index - Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='Close Price', color='blue')
plt.plot(df.index, df['Open'], label='Open Price', color='green')
plt.title('Nifty IT Index -  Price Over Time')
plt.xlabel('Date')
plt.ylabel(' Price')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.bar(df.index, df['Close'], color='orange')
plt.title('Nifty IT Index - Bar Chart of Close Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.hist(df['Close'], bins=20, color='green', alpha=0.7)
plt.title('Nifty IT Index - Histogram of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot(df['Close'], vert=False)
plt.title('Nifty IT Index - Box Plot of Close Prices')
plt.xlabel('Close Price')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(df.index, df['Close'], color='purple', alpha=0.5)
plt.title('Nifty IT Index - Scatter Plot of Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
